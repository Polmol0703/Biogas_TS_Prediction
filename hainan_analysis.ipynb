# Preliminary analysis of Hainan Data

```python
# import libraries
#IS AKSHAT ACTUALLY ON THIS TEAM???

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from string import ascii_letters
%matplotlib inline
```

```python
# read the Hainan file

hainan = pd.read_csv('hainan.csv')
hainan = hainan.drop('day_2', axis = 1) # drop day_2 column
hainan.head()
hainan_df = hainan

hainan_out = hainan[['raw_biogas_m3', 'biocng_prod_m3', 'biocng_sold_m3', 'lqd_fert_prod_t', 'lqd_fert_sold_t', 'solid_fert_prod_t', 'solid_fert_sold_t']]
hainan_input = hainan[['manure_inp_t', 'bagasse_1_inp_t', 'lees_ferm', 'fish_waste_inp_t', 'alcowaste_inp_t', 'cassava_inp_t', 'grass_inp_t', 'tea_waste_inp_t', 'other_inp_t', 'tot_inp_t', '50perc_naoh_kg', 'fecl2_kg', 'pam_kg', 'defoamer_kg', 'water_m3', 'tot_elec_cons_kwh', 'diesel_cons_lt']]
```

```python
X = hainan_out
Y = hainan_input

print(X.shape, Y.shape)

from sklearn.model_selection import train_test_split

X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.15)

print(X_train.shape, Y_train.shape)
print(X_val.shape, Y_val.shape)

from sklearn.linear_model import LinearRegression
linearreg= LinearRegression()
linearreg.fit(X_train, Y_train)

print('Coefficients: \n', linearreg.coef_)

print('Intercepts: \n', linearreg.intercept_)

acc_lin = linearreg.score(X_val, Y_val) # evaluate
acc_lin

print('Accuracy: \n', acc_lin)
```

```python
# Compute the correlation matrix
hainan_df = hainan.drop('year', axis = 1).drop('day', axis = 1).drop('month_num', axis = 1).drop('day_num', axis = 1).drop('month', axis = 1)

corr = hainan_df.corr()

# The cells that went up past the initial row are automatically deleted.

time_shift = 15

hainan_df.iloc[:,0:7] = hainan_df.iloc[:,0:7].shift(-time_shift)
# print(hainan_df.columns.values)
# print(hainan_df.tail())

# Drop the rows at the bottom
hainan_df.drop(hainan_df.tail(time_shift).index,inplace=True)

corr_timeshift = hainan_df.corr()
# print(corr_input)

corr_compare = corr_timeshift - corr

# Generate a mask for the upper triangle
mask_input = np.zeros_like(corr, dtype=np.bool)
mask_input[np.triu_indices_from(mask_input)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask_input, cmap=cmap, vmax=0.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

# Generate a mask for the upper triangle
mask_input2 = np.zeros_like(corr_timeshift, dtype=np.bool)
mask_input2[np.triu_indices_from(mask_input2)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap2 = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_timeshift, mask=mask_input, cmap=cmap2, vmax=0.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

# Generate a mask for the upper triangle
mask_input3 = np.zeros_like(corr_compare, dtype=np.bool)
mask_input3[np.triu_indices_from(mask_input3)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap3 = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_compare, mask=mask_input, cmap=cmap3, vmax=0.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
