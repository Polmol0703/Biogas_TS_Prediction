Biogas fermentation is a series of processes where microorganisms break down biodegradable materials (usually in the absence of oxygen), thus it is often attributed as anaerobic digestion. The break-down of these biodegradable materials (such as biomass) will produce simpler molecules, where some of these products are in the form of bio-gas. The detailed mechanism of how biogas fermentation occurs depends on the microorganisms involved, as well as types of feedstock (biomass) used and operating conditions (temperature, pH, etc). In general, we can describe the mechanism into four key biological and chemical stages: hydrolysis, acidogenesis, acetogenesis, methanogenesis
